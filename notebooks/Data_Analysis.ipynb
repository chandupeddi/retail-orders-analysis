As we know we cannot use the zip file we need to extract the zip file, so, I imported zipfile library and extracted the orders.csv.zip file to orders.csv file and placed in same location

```python
import zipfile
zip_ref = zipfile.ZipFile('orders.csv.zip')
zip_ref.extractall() 
zip_ref.close()
```

After file got extracted, I imported pandas library and loaded the csv file 

```python
import pandas as pd 
df = pd.read_csv('orders.csv')
```

After got loaded I checked the content in the file using head function to see what columns and data is present in the file 

```python
df.head(20)
```

In top 20 rows all the data looks good only the data in 'Ship Mode' column  have some missing data so first, I need to work on that missing data 

```python
df['Ship Mode'].unique()
```

For that I used unique function in that column and seen what values are there in the column 

As I can see I need to deal with some null values rows, some rows who has the value as 'Not Available', and some rows having value 'unknown'.The rest of the values are good.

To make it simple I changing the 'Not Available' and 'unknown' values into null values so we can only need to deal with null values. For that while loading file I making a extra correction by saying consider this values as null values 

```python
df = pd.read_csv('orders.csv',na_values = ['Not Available','unknown'])
```

By passing this argument, I turned those values into null values 

```python
df['Ship Mode'].unique()
```

Second thing, I need to handle is column names, as I seen them some of column names have space between them and some have captial letters in middle so I changing every column name by changing every column to lowercase and replacing spaces with underscore 

I am not using general method by writing every column and assigning it instead using functions 

```python
#General method 
#df.rename(columns = {'Order Id':'order_id','Country':'country'})
#Using functions
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
df.head(5)
```

Now the data looks good and columns look nice 

In the dataset we had cost of the item and MRP of the item and discount perecent gave for the item but it doesn't have at what price the item got sold after discount and how much discount gave and what is profit gain so we need to do some statstics to find those new values using given values and create three new columns as 'discount_price', 'sale_price',and profit.

```python
df['discount_price']= df['list_price']*df['discount_percent']*0.01
df['sale_price'] = df['list_price'] - df['discount_price']
df['profit'] = df['sale_price'] - df['cost_price']
```

```python
df.head(5)
```

I got new three columns and respected values, so now I checking each column data type so I get an idea is there any types of changes needed or all are correctly formatted 

```python
df.dtypes
```

All are correctly formated except 'order_date' column as I can see it data type was in 'object' but it should be in date data type so, I am converting it data type into date.

```python
df['order_date'] = pd.to_datetime(df['order_date'],format = "%Y-%m-%d")
```

Using pandas function 'to_datetime' and keeping exact format as in the values in dataset so it recongise and changed the entire column datatype

```python
df.dtypes
```

And coming to dataframe , I no longer needed cost_price, list_price and discount_percent columns in my dataframe for my further analysis so I dropping those columns from my dataframe 

```python
df.drop(columns=['list_price','cost_price','discount_percent'],inplace = True)
```
